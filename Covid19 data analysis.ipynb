```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style 
style.use('ggplot')               # create grid structure in background of matplotlib (see where we plot the bar graph)

%matplotlib inline

 
import plotly                       #library of data visualization
import plotly.express as px         # express is wrapper of plotly library
import plotly.graph_objects as go 
plt.rcParams['figure.figsize']=17,8
import cufflinks as cf               #Cufflinks connects the Pandas dataframe with Plotly to create visualizations directly from Pandas.
import plotly.offline as pyo
from plotly.offline import init_notebook_mode,plot,iplot

import folium
```

```python
df=pd.read_excel("Covid cases in India.xlsx")
df
```

```python
df.drop(['S. No.'],axis=1,inplace=True)
df
```

```python
df['Total Cases']=df['Total Confirmed cases (Indian National)']+df['Total Confirmed cases ( Foreign National )']
```

```python
df
```

```python
overall_total_cases=df['Total Cases'].sum()
print('The total number of cases is',overall_total_cases)
```

```python
df['Active Cases']=df['Total Cases']-(df['Death']+df['Cured'])
```

```python
df
```

```python
df.style.background_gradient(cmap='Reds')
```

```python
Total_Active_Cases=df.groupby('Name of State / UT')['Active Cases'].sum().sort_values(ascending=False).to_frame()
```

```python
Indian_Cord=pd.read_excel('Indian Coordinates.xlsx')
Indian_Cord
```

```python
df_full=pd.merge(Indian_Cord,df,on='Name of State / UT')
df_full
```

```python
map=folium.Map(location=[20,70],zoom_start=4,tiles='Stamenterrain')

for lat,long,value,name in zip(df_full['Latitude'],df_full['Longitude'],df_full['Total Cases'],df_full['Name of State / UT']):
    folium.CircleMarker([lat,long],radius=value*0.8,popup=('<strong>State</strong> : '+str(name).capitalize()+'<br>''<strong>Total Cases</strong> : '+str(value)+'<br>'),color='red',fill_color='red',fill_opacity=0.3).add_to(map)
```

```python
df.query('Country=="India"')
```

```python
df.groupby('Date').sum()
```

```python
confirmed=df.groupby('Date').sum()['Confirmed'].reset_index()
```

```python
confirmed
```

```python
deaths=df.groupby('Date').sum()['Deaths'].reset_index()
deaths
```

```python
recovered=df.groupby('Date').sum()['Recovered'].reset_index()
recovered
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=confirmed['Date'],y=confirmed['Confirmed'],mode='lines+markers',name='Confirmed',line=dict(color='blue',width=3)))
fig.add_trace(go.Scatter(x=deaths['Date'],y=deaths['Deaths'],mode='lines+markers',name='Deaths',line=dict(color='red',width=3)))
fig.add_trace(go.Scatter(x=recovered['Date'],y=recovered['Recovered'],mode='lines+markers',name='Recovered',line=dict(color='green',width=3)))
```
